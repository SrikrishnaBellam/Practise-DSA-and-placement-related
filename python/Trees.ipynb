```python
class Node:
    def __init__(self, key):
        self.left = None  # Left child
        self.right = None  # Right child
        self.value = key  # Node value

class BST:
    def __init__(self):
        self.root = None  # Root of the BST

    # Function to insert a key into the BST
    def insert(self, key):
        if key <= 0:
            return  # Only positive elements are allowed
        if self.root is None:
            self.root = Node(key)  # Create root if BST is empty
        else:
            self._insert(self.root, key)

    # Helper function to insert a key into the BST
    def _insert(self, root, key):
        if key < root.value:
            if root.left is None:
                root.left = Node(key)  # Insert as left child
            else:
                self._insert(root.left, key)
        else:
            if root.right is None:
                root.right = Node(key)  # Insert as right child
            else:
                self._insert(root.right, key)

    # Function to find the minimum value in the BST
    def find_min(self):
        if self.root is None:
            return None
        return self._find_min(self.root)

    # Helper function to find the minimum value in the BST
    def _find_min(self, root):
        current = root
        while current.left is not None:
            current = current.left  # Traverse left to find the min value
        return current.value

    # Function to find the maximum value in the BST
    def find_max(self):
        if self.root is None:
            return None
        return self._find_max(self.root)

    # Helper function to find the maximum value in the BST
    def _find_max(self, root):
        current = root
        while current.right is not None:
            current = current.right  # Traverse right to find the max value
        return current.value

    # Function to compute the height of the BST
    def height(self, node):
        if node is None:
            return 0
        else:
            left_height = self.height(node.left)
            right_height = self.height(node.right)
            return max(left_height, right_height) + 1

    # Function to get the height of the BST
    def get_height(self):
        return self.height(self.root)

    # Function to compute the diameter of the BST
    def diameter(self):
        if self.root is None:
            return 0
        return self._diameter(self.root)

    # Helper function to compute the diameter of the BST
    def _diameter(self, root):
        if root is None:
            return 0

        left_height = self.height(root.left)
        right_height = self.height(root.right)

        left_diameter = self._diameter(root.left)
        right_diameter = self._diameter(root.right)

        # Diameter is the max of left_diameter, right_diameter, or the path through the root
        return max(left_height + right_height + 1, max(left_diameter, right_diameter))

    # Function to perform inorder traversal
    def inorder(self):
        res = []
        self._inorder(self.root, res)
        return res

    # Helper function for inorder traversal
    def _inorder(self, root, res):
        if root:
            self._inorder(root.left, res)
            res.append(root.value)
            self._inorder(root.right, res)

    # Function to perform preorder traversal
    def preorder(self):
        res = []
        self._preorder(self.root, res)
        return res

    # Helper function for preorder traversal
    def _preorder(self, root, res):
        if root:
            res.append(root.value)
            self._preorder(root.left, res)
            self._preorder(root.right, res)

    # Function to perform postorder traversal
    def postorder(self):
        res = []
        self._postorder(self.root, res)
        return res

    # Helper function for postorder traversal
    def _postorder(self, root, res):
        if root:
            self._postorder(root.left, res)
            self._postorder(root.right, res)
            res.append(root.value)

    # Function to perform level order traversal
    def level_order(self):
        res = []
        if self.root is None:
            return res
        queue = [self.root]
        while queue:
            node = queue.pop(0)
            res.append(node.value)
            if node.left is not None:
                queue.append(node.left)
            if node.right is not None:
                queue.append(node.right)
        return res

    # Function to perform vertical order traversal
    def vertical_order(self):
        if not self.root:
            return []
        from collections import defaultdict, deque
        column_table = defaultdict(list)
        queue = deque([(self.root, 0)])

        while queue:
            node, column = queue.popleft()
            if node is not None:
                column_table[column].append(node.value)
                queue.append((node.left, column - 1))
                queue.append((node.right, column + 1))

        return [column_table[x] for x in sorted(column_table.keys())]

    # Function to compute the sum of all nodes in the BST
    def sum_of_nodes(self):
        return self._sum_of_nodes(self.root)

    # Helper function to compute the sum of all nodes in the BST
    def _sum_of_nodes(self, node):
        if node is None:
            return 0
        return node.value + self._sum_of_nodes(node.left) + self._sum_of_nodes(node.right)

    # Function to search for a key in the BST
    def search(self, key):
        return self._search(self.root, key)

    # Helper function to search for a key in the BST
    def _search(self, root, key):
        if root is None or root.value == key:
            return root
        if key < root.value:
            return self._search(root.left, key)
        return self._search(root.right, key)

# Example usage
bst = BST()
values = [10, 5, 1, 7, 40, 50]
for val in values:
    bst.insert(val)

print("Inorder Traversal:", bst.inorder())
print("Preorder Traversal:", bst.preorder())
print("Postorder Traversal:", bst.postorder())
print("Level Order Traversal:", bst.level_order())
print("Vertical Order Traversal:", bst.vertical_order())
print("Min Value:", bst.find_min())
print("Max Value:", bst.find_max())
print("Height:", bst.get_height())
print("Diameter:", bst.diameter())
print("Sum of Nodes:", bst.sum_of_nodes())
print("Search 7:", bst.search(7) is not None)
print("Search 100:", bst.search(100) is not None)
```

```python
def delete(node, key):
    if node is None:
        return node
    if key<node.key:
        node.left = delete(node.left, key)
    elif key > node.key:
        node.right = delete(node.right, key)
    else:
        if node.left is None:
            return node.right
        elif node.right is None:
            return node.left
        
        temp = find_min(node.right)
        node.key = temp.key
        node.right = delete(node.right, temp.key)

def find_min(node):
    curr = node
    while curr.left is not None:
        curr = curr.left
        return curr
```

```python
def find_max(node):
    curr = node
    while curr.right is not None:
        curr = curr.right
        return curr
```

```python
def find_height(node):
    if node is None:
        return -1
    left_height =find_height(node.left)
    right_height =find_height(node.right)

    return max(left_height, right_height) + 1
```
